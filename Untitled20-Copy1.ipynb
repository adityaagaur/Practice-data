```python
def outlier_capping(x):
    x=x.clip(upper=x.quantile(0.99))
    x=x.clip(lower=x.quantile(0.01))
    return(x)
```

```python
data1=data.apply(outlier_capping)
```

```python
sns.pairplot(data1)
```

```python
plt.figure(figsize=(10,8))
df=data1.corr()
sns.heatmap(df,cmap="rocket_r",annot= True)
plt.show()
```

```python
x=data1.drop(["default"],axis=1)
y=data1["default"]
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train ,y_test=train_test_split(x,y, test_size=0.25,random_state=88)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log0=LogisticRegression()
```

```python
log0.fit(x_train,y_train)
```

```python
print("Train Acuracy:",log0.score(x_train,y_train))
```

```python
print("Test accuracy :",log0.score(x_test,y_test))
```

```python
y_train
```

```python
pred_train=log0.predict(x_train)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_train,pred_train)
```

```python
print(metrics.classification_report(y_train,pred_train))
```
